```python
import pandas as pd
import numpy as np
```

```python
INVERST_COST = 500
df = pd.read_csv('../data/stocks.csv', index_col='stocks')

df['return'] = df['cost'] * df['roi'] / 100

stock_cost_df = df.drop(columns='roi')
stock_return_df = df.drop(columns='cost')

df
```

```python
stock_cost_dict = stock_cost_df.to_dict()['cost']
stock_return_dict = stock_return_df.to_dict()['return']
```

```python
stock_return_dict
```

```python
stock_cost_dict
```

```python
def generate_stock_combinations(stock_names, r, current_combination, stock_combinations):
    if r == 0:
        stock_combinations.append(current_combination)
        return
    for i in range(len(stock_names)):
        new_combination = current_combination + [stock_names[i]]
        generate_stock_combinations(
            stock_names[i+1:], r-1, new_combination, stock_combinations)
```

```python
stocks = df.index
stocks
```

```python
stock_combinations = []
for r in range(1, len(stocks) + 1):
    generate_stock_combinations(stocks, r, [], stock_combinations)
```

```python
stock_combinations
```

```python
#stock_combinations_array = np.array(stock_combinations)
```

```python
column_values = np.array(list(range(1, 21)))
column_values
```

```python
comb_df = pd.DataFrame(data=stock_combinations,
                       columns=column_values)
comb_df.index += 1
```

```python
comb_df.fillna(0,inplace=True)
comb_df
```

```python
df = df.T
df
```

```python
def replace_by_return(x): return stock_return_dict.get(x, x)
def replace_by_cost(x): return stock_cost_dict.get(x, x)


return_df = comb_df.applymap(replace_by_return)
cost_df = comb_df.applymap(replace_by_cost)
```

```python
return_df
```

```python
cost_df
```

```python
cost_df['sum'] = cost_df.sum(axis=1, skipna=False)
cost_df
```

```python
return_df.replace(0, np.nan, inplace=True)
return_df['sum'] = return_df.sum(axis=1, skipna=True)
return_df
```

```python
comb_df['total return'] = return_df['sum']
comb_df['cost'] = cost_df['sum']
comb_df
```

```python
comb_df.replace(0, np.nan, inplace=True)
comb_df
```

```python
budget_df = comb_df.loc[comb_df['cost'] <= INVERST_COST]
budget_df
```

```python
comb_performance = best_stocks_comb['total return'] / best_stocks_comb['cost'] * 100
```
